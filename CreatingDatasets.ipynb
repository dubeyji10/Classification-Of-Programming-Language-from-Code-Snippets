### function for reading file at a directory

```
def read_this(path):
    with open(path,'r',encoding='utf-8') as f:
        data = f.read()
    return data
```

```
%%time

for i in dataset_dict.keys():
  directory = path+i
  if os.path.isdir(directory):
    print('accessing :',directory)
    files_in_directory = listdir(directory)
    print('\ttotal files read:',len(files_in_directory))
    # for j in range(2000):
    for j in int_index1:
      sample_read0 = directory+''.join('/'+''.join(files_in_directory[j]))
#             print('*'*50)
#             print('\tcontent :\n ',read_this(sample_read0))
#             print('*'*50)
      dataset_dict[i].append(read_this(sample_read0))
    print('\t <- {} files appended ->'.format(len(dataset_dict[i])))
            
    print('-'*100)
```

## Append dictionary to dataframe

```
%%time
print('takes around 2min for completion')
counter_i= 0
for i in dataset_dict:
    print('appending for ',i)
    for j in range(len(dataset_dict[i])):
        row = int(counter_i*len(dataset_dict[i]))+j
        empty_df.loc[row] = [i] + [dataset_dict[i][j]]
        # print('added data at row=',row)
    counter_i+=1
    print('->counter_i = ',counter_i)
    print('-'*70)

print("\n\n ------------------------------ DONE ------------------------------")
```
